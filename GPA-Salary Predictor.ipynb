You would like to fit a linear regression model (shown above as the blue straight line) through these  points, so you can then predict salary for other grade point averages - say, a theeta with 3.9 GPA.


Create 1D numpy arrays to store features and target variables.

```python
# x_train is the input variable (GPA)
# y_train is the output variable (Salary)

x_train = np.array([2.5,2.55,2.6,2.69,2.7,2.8,2.9,3.0])
y_train = np.array([300.0,500.0,600.0,3000.0,650.0,900.0,1000.0,1100.0])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

### Plotting the data

- The function arguments `marker` and `c` show the points as red crosses (the default is blue dots).

You can use other functions in the `matplotlib` library to set the title and labels to display.


```python
#plot the data points
plt.scatter(x_train,y_train,marker = 'x', c = 'r')

#Set title
plt.title("GPA-Salary Correlation")

#Set x axis label
plt.xlabel('GPA')

#Set y axis label
plt.ylabel('Salary (in $100)')
```

## Model function


The model function for linear regression (which is a function that maps from `x` to `y`) is represented as 

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

The formula above is how you can represent straight lines - different values of $w$ and $b$ give you different straight lines on the plot.

Now, let's compute the value of  ùëìùë§,ùëè(ùë•(ùëñ))
  for our data. 
You can calculate the function output in a for loop as shown in the compute_model_output function below.

Note: The argument description (ndarray (m,)) describes a Numpy n-dimensional array of shape (m,). (scalar) describes an argument without dimensions, just a magnitude.
Note: np.zero(n) will return a one-dimensional numpy array with  ùëõ entries

```python
def compute_model_output(x_train, w, b):
    m = len(x_train)
    f_x = np.zeros(m)
    # print(f"f(X) = {f_x}")

    for i in range(m):
        f_x[i] = w * x_train[i] + b
    
    return f_x
```

Now, let us call the model function.

```python
w = 200
b = 200

tmp_f_x = compute_model_output(x_train, w, b,)
```

Now, let's plot the output(the straight line :)

```python
# Plot our model prediction
plt.plot(x_train, tmp_f_x, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

#Set title
plt.title("GPA-Salary Correlation")

#Set x axis label
plt.xlabel('GPA')

#Set y axis label
plt.ylabel('Salary (in $100)')

plt.legend()
plt.show()
```

As you can see, setting  ùë§= 200 and  ùëè= 200 results in a line that best fits our data.
If you want to play around with w and b, you can change the values in line 60 :)
